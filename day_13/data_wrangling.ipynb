Exploratory data analsis is a method used to analyzee and sumarize datasets. In python you can use the panda libraray to perform eda on a dataset.


```python
#import libraraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data set
df=sns.load_dataset("titanic")
# analyse and sumarize the dataset
```

```python
#sumarize data
df.describe()
```

```python
#find null values
df.isnull()
```

```python
df.isnull().sum()
```

```python
#to calculate in percentage null values PEMDAS rule
df.isnull().sum()/len(df)*100
```

```python
#finding unique values
# Thumb rule: if a data have >10uniques values then it is numeric,float or int.
# if a data have <10 unique values then it is categorical

df.nunique()
```

```python
df.columns
```

```python
#Visualize the nan value in heatmap
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap="viridis")
```

```python
df.drop("deck",axis=1,inplace=True)
```

```python
#shape is an attribute, so it is accessed using dot notation and not paranthesis.
df.shape
```

```python
#visualize and explore
sns.countplot(data=df,x="survived",hue='sex',)
# for p in plt.gca().patches: plt.gca().annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')
# plt.show()
```

```python
df.columns
sns.barplot(x="survived",y="fare",hue="who",data=df)
#x-axis categorical and y-axis continuous or numerical 
```

```python
sns.countplot(x="pclass",data=df)
#now analyse and sumarize.
#analyse=1 2 and 3 class passengers count
#summarize: 3class >people
#in countplot only give categorical variable automatically it will
#make bar plot 
```

```python
#check Data distribution, use histogram chart.
#Bins means the number of intervals in the histogram. The default is 10.
sns.histplot(x="age",data=df,bins=30)
```

```python
df['age'].hist(bins=30,color='darkred',alpha=0.75)
#alpha means opasisty/ adjust transparency. 
#hist plot works only both sides of axis should be numerical,
```

```python
sns.boxplot(data=df,x='survived',hue='who',y="age")
#boxplot works only oneside numerical and otherside categorical
# bins will become if more data points
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='pclass')
plt.ylim(0,300)
plt.show()
#here graph is not showing perfectly and clear, so we will normalization of data.
```

# ___PAIRPLOT___

```python
sns.pairplot(df,hue='survived')
```

```python
#Stacked barchart/plot teh wali baryani
df.groupby(["pclass","who"])['fare'].mean().unstack().plot(kind='bar',stacked=True)
```

```python
#SWARM
sns.swarmplot(data=df,x="survived",y="age",hue="who")
plt.show()
```

```python
#point plot
sns.catplot(x='pclass',y='fare',hue='embarked',kind='point',data=df)
#when dr.aamar discuss statistics that time will learn in detail.
```

```python
#LM plot. to check relation of two numeric variables
sns.lmplot(data=df, x='fare',y='age')
```

```python
sns.lmplot(data=df,x="age",y="fare",hue='who')
```

```python
filtered_df=df[df["fare"]<300]
sns.lmplot(data=filtered_df,x="age",y="fare",hue='who')
```

```python
#how to add a column and make a age group
sns.histplot(df,x="age")
```

```python
df["age_groups"]=pd.cut(df['age'],[0,10,18,30,50,80],
                        labels=['Kids','Child','Youth','Middle_age','Old'])
```

```python
df.head()
```

```python
df.columns
```

```python
df.groupby("age_groups")['survived'].mean().plot(kind='bar')
```

```python
sns.pairplot(df,hue='survived',vars=['age','fare','pclass'])
```
