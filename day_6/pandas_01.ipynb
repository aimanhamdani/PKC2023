#16. SPLIT A STRING INTO MULTIPLE COLUMNS

```python
df=pd.DataFrame({'name':["Aiman Aali","Mohammad Mehdi", "Najam-us Saqib"],
                 "location":["Qatar,Doha","Pakistan,Talagang","Paksitan,Talagang1"]
})
df
```

```python
df[["First_Name","Last_Name"]]=df.name.str.split(" ",expand=True)
df
```

```python
df["hello"]=[1,2,3]
df
```

```python
df[["county","city"]]=df.location.str.split(",",expand=True)
df
```

# RESHAPING A DATAFRAME

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fasla= pd.DataFrame([["00000",100,200,300],["11111",400,500,600],["22222",700,800,900]],columns=["zip_code","factory","warehouse", "retail"])
print(fasla.head())
print(fasla.columns)
```

```python
melted_fasla=fasla.melt(id_vars="zip_code",value_name='distance',var_name='location')
melted_fasla
```

```python
sns.barplot(x="zip_code", y="distance", hue="location",data=melted_fasla)
```
