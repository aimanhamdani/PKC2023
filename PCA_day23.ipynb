```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Sample data (marks in various subjects)
X = np.array([
    [85, 90, 78, 65, 70],
    [80, 85, 85, 70, 75],
    [78, 88, 80, 72, 77],
    [70, 75, 85, 75, 80],
    [65, 70, 78, 80, 85]
])
```

```python
X
```

```python

# Apply PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)
```

```python
principalComponents
```

```python

# Plot PCA results
plt.figure()
plt.scatter(principalComponents[:, 0], principalComponents[:, 1], edgecolor='k', s=100)
for i in range(principalComponents.shape[0]):
    plt.text(principalComponents[i, 0], principalComponents[i, 1], f'Student {i+1}')

plt.title('PCA of Students\' Marks')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```
